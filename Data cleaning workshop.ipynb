```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/SP112498/Downloads/squirrel_data.csv')
```

```python
numeric_cols=df.select_dtypes(include=['number']).columns 
print(numeric_cols)
```

```python
non_numeric_cols=df.select_dtypes(exclude=['number']).columns
print(non_numeric_cols)
```

```python
num_missing=df.isna().sum()
print(num_missing)
```

```python
pct_missing=df.isna().mean() #empty parenthesis means to execute, 
print(pct_missing*100)
```

```python
#shows missing data bu columns
plt.figure(figsize=(10,8))
cols=df.columns
colors=['blue','yellow']
sns.heatmap(df[cols].isna(), cmap=sns.color_palette(colors))
```

```python
#shows missing columns by rows
plt.figure(figsize=(10,8))
missing_by_row=df.isna().sum(axis='columns')
missing_by_row.hist(bins=50)
```

```python
df_copy1=df.copy()
pct_missing[pct_missing>0.3]
```

```python
df_less_missing_cols=df_copy1.loc[:,pct_missing<=0.3]#selecting columns less than or equal to 30%
df_less_missing_cols.shape #originally it had 31 columns but as 5 columns more than 30% missing data we droppd them out
```

```python
print(df_less_missing_cols)
```

```python
# () to call a function >> parenthesis
# [] is more like to create a list
df_copy2=df.copy()
df_less_missing_rows=df_copy2[missing_by_row<6]
df_less_missing_rows.shape
```

```python
df_copy3=df.copy()
df_copy3[non_numeric_cols]=df_copy3[non_numeric_cols].fillna('_MISSING_')
df_copy3.head()
```

```python
df_copy4=df.copy()
most_freq=df_copy4[non_numeric_cols].describe().loc['top']
most_freq
```

```python
df_copy4[non_numeric_cols]=df_copy4[non_numeric_cols].fillna(most_freq)
df_copy4.head()
```

```python
output=os.path.join('C:/Users/SP112498/Downloads/squirrel_data_cleaned.csv')
df_copy4.to_csv(output)
```
